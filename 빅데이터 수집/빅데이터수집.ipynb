```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import csv
```

```python
# 1.위키디피아 사이트에서 빅데이터를 검색한 후, 정의에 해당하는 텍스트내용을 콘솔에 출력
driver = webdriver.Chrome()
url = 'https://ko.wikipedia.org/wiki/'
driver.get(url)
time.sleep(1)
driver.find_element(By.XPATH,'//*[@id="p-search"]/a').click()
time.sleep(1)
driver.find_element(By.XPATH,'//*[@id="searchform"]/div/div/div[1]/input').send_keys('빅데이터')
driver.find_element(By.XPATH,'//*[@id="searchform"]/div/div/div[1]/input').send_keys(Keys.ENTER)
time.sleep(2)
txt=driver.find_element(By.XPATH,'//*[@id="mw-content-text"]/div[1]/p[6]').text
print(txt)
```

```python
# 2. 국민 동의 청원사이트에서 동의 진행중인 청원 내용을 수집하여 '국민동의.txt'파일로 저장
driver = webdriver.Chrome()
url = 'https://petitions.assembly.go.kr/'
driver.get(url)
time.sleep(2)
driver.find_element(By.XPATH,'//*[@id="header"]/div/div[2]/nav/div[2]/div[1]').click()
time.sleep(1)
driver.find_element(By.XPATH,'//*[@id="header"]/div/div[3]/ul/li[2]/ul/li[1]/a').click()
time.sleep(2)
page_cnt = int(driver.find_element(By.XPATH,'//*[@id="contentsbody"]/div/div/div[4]/div[2]/ul/li[9]/a').text)

ass_file = open('data/assembly.txt','w',encoding='utf-8')

for i in range(page_cnt):
    pclick = len(driver.find_elements(By.XPATH,'//*[@id="contentsbody"]/div/div/div[4]/div[2]/ul/li'))-1
    driver.find_element(By.XPATH,'//*[@id="contentsbody"]/div/div/div[4]/div[2]/ul/li['+str(pclick)+']/a').click()
    time.sleep(1)
    list_cnt = driver.find_elements(By.XPATH,'//*[@id="contentsbody"]/div/div/div[4]/div[1]/div[2]/ul/li')
    time.sleep(1)
    for j in range(1,len(list_cnt)+1):
        title = driver.find_element(By.XPATH,'//*[@id="contentsbody"]/div/div/div[4]/div[1]/div[2]/ul/li['+str(j)+']/div/a/dl/dd').text
        ass_file.write(title+"\n")
        print(title)
        
    time.sleep(1)
ass_file.close()
driver.close()
```

```python
# 3. 오피넷 사이트에서 대전 서구의 주유소 보통 휘발유가격을 가져와 엑셀파일로 작성
driver = webdriver.Chrome()
url = 'https://www.opinet.co.kr/user/main/mainView.do'
driver.get(url)
time.sleep(2)
driver.find_element(By.XPATH,'//*[@id="quick_ul"]/li[2]/a').click()
time.sleep(2)

driver.find_element(By.XPATH,'//*[@id="SIDO_NM0"]/option[7]').click()
driver.find_element(By.XPATH,'//*[@id="SIGUNGU_NM0"]/option[4]').click()
time.sleep(1)
driver.find_element(By.XPATH,'//*[@id="searRgSelect"]').click()
time.sleep(1)

file = open('data/oil_sugu.csv','w',newline='',encoding='cp949')
writer = csv.writer(file)

rows = driver.find_elements(By.XPATH,'//*[@id="body1"]/tr')

for i in range(1,len(rows)+1):
    name = driver.find_element(By.XPATH,'//*[@id="body1"]/tr['+str(i)+']/td[1]/a').text
    g_price = driver.find_element(By.XPATH,'//*[@id="body1"]/tr['+str(i)+']/td[2]').text
    v_price = driver.find_element(By.XPATH,'//*[@id="body1"]/tr['+str(i)+']/td[3]').text
    print(name,' | ',g_price,' | ',v_price)
    writer.writerow([name,g_price,v_price])
file.close()
driver.close()
```
